```
train_dir = '/tmp/training/'
import os
train_dis_dir = os.path.join(train_dir,'Disappointed')
train_int_dir = os.path.join(train_dir,'interested')
train_neutral_dir = os.path.join(train_dir,'neutral') 
test_dir ='/tmp/testing/'
```

```
dis_files =os.listdir(train_dis_dir)
int_files=os.listdir(train_int_dir)
neutral_files = os.listdir(train_neutral_dir)
```

```
pic_index = 2
next_dis = [os.path.join(train_dis_dir,fname) for fname in dis_files[pic_index-2:pic_index]]
next_int =[os.path.join(train_int_dir,fname) for fname in int_files[pic_index-2:pic_index]]
next_neutral = [os.path.join(train_neutral_dir,fname) for fname in neutral_files[pic_index-2:pic_index]]
```

```
for i,img_path in enumerate(next_dis+next_int+next_neutral):
  img= mpimg.imread(img_path)
  plt.imshow(img)
  plt.axis('Off')
  plt.show()
```

```
train_datagen = ImageDataGenerator(rescale=1.0/255.,rotation_range=40,zoom_range=0.2,shear_range=0.2,width_shift_range=0.2,height_shift_range=0.2,fill_mode='nearest',horizontal_flip=True)
```

```
test_datagen = ImageDataGenerator(rescale=1.0/255.)
```

```
train_data_generator =train_datagen.flow_from_directory(train_dir,class_mode='categorical',target_size=(150,150),batch_size=32)
test_data_generator = test_datagen.flow_from_directory(test_dir,target_size=(150,150),class_mode='categorical',batch_size=32)
```

```

!wget --no-check-certificate \
    https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5 \
    -O /tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras import Model
local_weights_file = '/tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'
pre_trained_model = InceptionV3(input_shape=(150,150,3),include_top=False,weights=None)
pre_trained_model.load_weights(local_weights_file)
for layer in pre_trained_model.layers:
  layer.trainable = False
last_layer = pre_trained_model.get_layer('mixed9')
last_layer_output = last_layer.output
x = layers.Flatten()(last_layer_output)
x= layers.Dense(512,activation='relu')(x)
x= layers.Dropout(0.9)(x)
x= layers.Dense(3,activation='softmax')(x)
model = Model(pre_trained_model.input,x)
model.compile(loss='categorical_crossentropy',optimizer=RMSprop(lr=0.0001),metrics=['accuracy'] )



```

```
history = model.fit(train_data_generator,epochs=10,validation_data=test_data_generator)
```
